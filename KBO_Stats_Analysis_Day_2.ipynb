```python
# train set과 test set 분리
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
train_set.to_csv("KB0_24_train.csv", index = False)
test_set.to_csv("KB0_24_test.csv", index = False)
```

```python
# 예측 변수와 타깃값 분리
WAR = train_set.drop("WAR", axis=1)
WAR_labels = train_set["WAR"].copy()
```

```python
# StansdardScaler 사용하여 스케일링
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
WAR_scaled = std_scaler.fit_transform(WAR)
```

```python
WAR_scaled
```

```python
# 파이프라인 생성
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])
```

```python
# 훈련 시작
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(WAR_scaled, WAR_labels)
```

```python
WAR_predictions = lin_reg.predict(WAR_scaled)
WAR_predictions[:5]
```

```python
WAR_labels.iloc[:5].values
```

```python
# 정확도 판단
from sklearn.metrics import root_mean_squared_error

lin_rmse = root_mean_squared_error(WAR_labels, WAR_predictions)
lin_rmse
```

```python
# 결정트리로 훈련해보자.
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(WAR_scaled, WAR_labels)
```

```python
WAR_predictions = tree_reg.predict(WAR_scaled)
tree_rmse = root_mean_squared_error(WAR_labels, WAR_predictions)
tree_rmse
```

```python
# 오차가 없다고? 교차 검증으로 평가하기
from sklearn.model_selection import cross_val_score

scores = - cross_val_score(tree_reg, WAR_scaled, WAR_labels,
                         scoring="neg_mean_squared_error", cv=10)
```

```python
pd.Series(scores).describe()
```

```python
# 랜덤 포레스트 사용하기
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(random_state=42)
forest_reg.fit(WAR_scaled, WAR_labels)
```

```python
forest_rmses = - cross_val_score(forest_reg, WAR_scaled, WAR_labels,
                         scoring="neg_root_mean_squared_error", cv=10)
```

```python
pd.Series(forest_rmses).describe()
```

```python
# 특성의 중요도 추출
final_model = forest_reg
feature_importances = final_model.feature_importances_
feature_importances.round(2)
```

```python
sorted(zip(feature_importances, WAR.columns), reverse=True)
```

```python
# test set으로 평가하기
WAR_test = test_set.drop("WAR", axis=1)
WAR_labels_test = test_set["WAR"].copy()

WAR_test_scaled = std_scaler.transform(WAR_test)

final_predictions = final_model.predict(WAR_test_scaled)

final_rmse = root_mean_squared_error(WAR_labels_test, final_predictions)
final_rmse
```

```python
from scipy import stats
confidence = 0.95
squared_errors = (final_predictions - WAR_labels_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

```python
# 모델 저장
import joblib
joblib.dump(final_model, "KB0_24_WAR_prediction_model.pkl")
```
