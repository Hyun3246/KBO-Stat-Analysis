## NonLinear SVM & Hyperparameter RandomSearch

```python
# train set과 test set 분리
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# train set과 dev set 분리
train_set, dev_set = train_test_split(train_set, test_size=0.33, random_state=42)
```

```python
R = train_set["R"].to_numpy()
R = R.reshape(-1, 1)

RBI = train_set["RBI"].to_numpy()
RBI = RBI.reshape(-1, 1)

R_AND_RBI = np.concatenate((R, RBI), axis=1)

y = train_set["WAR"].copy().to_numpy()
y = y.reshape(-1, 1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint


full_pipeline = Pipeline([
    ("StandardScaler", StandardScaler()),
    ("SVR", SVR(kernel='poly')),
    ])

param_distribs = {'SVR__degree': randint(low=2, high=10),
                  'SVR__C': np.random.uniform(0.01, 0.1, size=10),
                  'SVR__epsilon': np.random.uniform(0.01, 0.1, size=10),
                  }

rnd_search = RandomizedSearchCV(full_pipeline, param_distributions=param_distribs, n_iter=10, cv=3,
                                 scoring='neg_mean_squared_error', random_state=42)

rnd_search.fit(R_AND_RBI, y)
```

```python
rnd_search.best_params_
```

```python
final_model = rnd_search.best_estimator_
```

```python
R_test = test_set["R"].to_numpy()
R_test = R_test.reshape(-1, 1)

RBI_test = test_set["RBI"].to_numpy()
RBI_test = RBI_test.reshape(-1, 1)

R_AND_RBI_test = np.concatenate((R_test, RBI_test), axis=1)
```

```python
from sklearn.metrics import root_mean_squared_error

y_test = test_set["WAR"].copy().to_numpy()
y_test = y_test.reshape(-1, 1)

final_prediction = final_model.predict(R_AND_RBI_test)
final_rmse = root_mean_squared_error(y_test, final_prediction)
final_rmse
```
